```python
import numpy as np
from HelperFunctions import LossFunctions
import NeuralNetworkLayer as NNL
import NeuralNetwork as NN
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn import metrics
from Optimizers import SGD
import time
import datetime
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
print(f'{X_train.shape=}')
print(f'{y_train.shape=}')
print(f'{X_test.shape=}')
print(f'{y_test.shape=}')
```

```python
X_train = X_train.reshape(X_train.shape[0], -1) / 255
X_test = X_test.reshape(X_test.shape[0], -1) / 255
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)

print(f'{X_train.shape=}')
print(f'{y_train.shape=}')
print(f'{X_test.shape=}')
print(f'{y_test.shape=}')
```

```python
no_epochs = 50
n = int(X_train.shape[0] / 50)
lr = 0.01
```

```python
start = time.time()
nn = NN.NeuralNetwork(X_train.shape[1])
nn.add(NNL.DenseLayer(8))
nn.add(NNL.RelUActivation())
nn.add(NNL.DenseLayer(y_train.shape[1]))
nn.add(NNL.SoftmaxActivation())

nn.describe()

epochs = no_epochs


batch_size = 1
losses = nn.fit(X_train[:n], y_train[:n], epochs,
                SGD(lr), batch_size=batch_size)

# 
y_preds = nn.predict(X_test)

# now pick the most likely outcome
pred = np.argmax(y_preds, axis=1)
y_compare = np.argmax(y_test, axis=1)
score = metrics.accuracy_score(y_compare, pred)
print("Test Accuracy: {}".format(score))
print(f'Total time: {datetime.timedelta(seconds=time.time() - start)}')
```
